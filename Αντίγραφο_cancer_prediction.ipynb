# **Cancer Prediction**

Dataset Information:

Target Variable (y):
- Diagnosis (M = malignant, B = benign)

Ten features (X) are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter)
2. texture (standard deviation of gray-scale values)
3. perimeter
4. area
5. smoothness (local variation in radius lengths)
6.  compactness (perimeter^2 / area - 1.0)
7. concavity (severity of concave portions of the contour)
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension (coastline approximation - 1)

For each characteristic three measures are given:

a. Mean

b. Standard error

c. Largest/ Worst

```python
# Step 3 : define target (y) and features (X)
```

```python
cancer.columns
```

```python
y = cancer['diagnosis']
```

```python
X = cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

cancer.columns: Τυπώνει τα ονόματα όλων των στηλών για να ξέρουμε πώς να τις καλέσουμε.

y (Target): Επιλέγουμε τη στήλη diagnosis. Αυτό είναι που θέλουμε να προβλέψουμε (M = Κακόηθες, B = Καλόηθες).

X (Features): Εδώ αφαιρούμε ό,τι δεν θέλουμε να χρησιμοποιήσει το μοντέλο:

diagnosis: Γιατί είναι η απάντηση!

id: Γιατί είναι απλώς ένας αριθμός μητρώου και δεν επηρεάζει το αν κάποιος έχει καρκίνο.

Unnamed: 32: Γιατί, όπως είδαμε στο info(), είναι μια στήλη "σκουπίδι" με κενές τιμές.

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

Τι κάνει: Χωρίζει τα δεδομένα μας σε δύο ομάδες:

Train (70%): Τα βιβλία που θα "διαβάσει" το μοντέλο για να μάθει.

Test (30%): Οι εξετάσεις που θα δώσει στο τέλος για να δούμε τι έμαθε.

random_state=2529: Αυτός είναι ο "σπόρος" της τυχαιότητας. Αν δεν τον βάζαμε, κάθε φορά που τρέχαμε τον κώδικα θα παίρναμε διαφορετικό διαχωρισμό και άρα ελαφρώς διαφορετικά αποτελέσματα. Με αυτό, εξασφαλίζουμε ότι το πείραμα είναι επαναλήψιμο.

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

Τι κάνει: Δημιουργεί ένα αεκπαίδευτο μοντέλο Λογιστικής Παλινδρόμησης 📉.

max_iter=5000: Η Λογιστική Παλινδρόμηση χρησιμοποιεί έναν επαναληπτικό αλγόριθμο για να βρει τη λύση. Η προεπιλογή είναι συχνά 100 επαναλήψεις, που μπορεί να μην αρκούν για πολύπλοκα δεδομένα. Αυξάνοντάς το σε 5000, δίνουμε στο μοντέλο άπλετο χρόνο να συγκλίνει στη βέλτιστη λύση.

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

Η καρδιά της Μηχανικής Μάθησης: Εδώ το μοντέλο μελετά τα X_train (χαρακτηριστικά) και τις αντίστοιχες απαντήσεις y_train για να βρει τα μαθηματικά μοτίβα που τα συνδέουν.

```python
model.intercept_
```

Αυτά είναι τα "σωθικά" της εξίσωσης που έμαθε το μοντέλο.

coef_: Δείχνει πόσο βαραίνει το κάθε χαρακτηριστικό (π.χ. η ακτίνα) στην τελική απόφαση.

intercept_: Είναι ο σταθερός όρος της εξίσωσης (το σημείο τομής).

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

Η στιγμή της κρίσης: Ζητάμε από το μοντέλο να μαντέψει τη διάγνωση για τους ασθενείς του X_test, τους οποίους δεν έχει δει ποτέ ξανά.

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

confusion_matrix: Μας δείχνει τα λάθη αναλυτικά (π.χ. πόσα κακοήθη τα πέρασε για καλοήθη).

accuracy_score: Το ποσοστό επιτυχίας (π.χ. 95%).

classification_report: Μια λεπτομερής αναφορά που μας δίνει την Ακρίβεια (Precision), την Ανάκληση (Recall) και το F1-score για κάθε κλάση ξεχωριστά. Είναι πολύ πιο χρήσιμο από το σκέτο accuracy σε ιατρικά θέματα.

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
